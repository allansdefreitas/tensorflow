```python
#Multi-Layer Perceptron using Low level API from TensorFlow
from sklearn import datasets
import tensorflow as tf
import numpy as np
```

```python
#https://archive.ics.uci.edu/ml/datasets/Iris
iris = datasets.load_iris()
#Attributes
#1. sepal length in cm
#2. sepal width in cm
#3. petal length in cm
#4. petal width in cm

#Class:
#-- Iris Setosa
#-- Iris Versicolour
#-- Iris Virginica
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
#Padronização: 
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)
X
```

```python
#One hot encoding: adjust the data format to network exit layer format: 3 columns equal to 3 exit neurons equals to 3 classes!
"""
[0]         [1][0][0]       Setosa
[1]    to   [0][1][0]       Versicolour
[2]         [0][0][1]       Virginica

"""
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder(categorical_features = [0])
#just lines: array format
y.shape
```

```python
y = y.reshape(-1, 1)

#lines and columns: matrix format: ok!
y.shape
```

```python
y = onehot.fit_transform(y).toarray()
y
```

```python
#Partition to trainning and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
#Mounting the NEURAL NETWORK ARCHITECTURE: input layer
neurons_input_layer = X.shape[1] #
neurons_input_layer
```

```python
#Mounting the NEURAL NETWORK ARCHITECTURE: hidden layer
#Formula to discover the number of neurons in hidden layer: (number of atts PLUS number of classes) DIVIDED BY 2
#np.ceil round a floating number to down
neurons_hidden_layer = int(np.ceil((X.shape[1] + y.shape[1]) / 2))
neurons_hidden_layer
```

```python
#Mounting the NEURAL NETWORK ARCHITECTURE: exit layer
neurons_exit_layer = y.shape[1]
neurons_exit_layer
```

```python
#Matrix of Weights
W = {'hidden': tf.Variable(tf.random_normal([neurons_input_layer, neurons_hidden_layer])),
     'exit': tf.Variable(tf.random_normal([neurons_hidden_layer, neurons_exit_layer]))}
```

```python
#Array of Biases
b = {'hidden': tf.Variable(tf.random_normal([neurons_hidden_layer])),
     'exit': tf.Variable(tf.random_normal([neurons_exit_layer]))}
```

```python
#Using placeholders: 

#None: It dont specificate the number of elements (objects),
#neurons_input_layer: number of features(columns, attributes)
xph = tf.placeholder('float', [None, neurons_input_layer])

#neurons_exit_layer: number of classes(i.e. 3): Setosa, Versicolour and Virginica
yph = tf.placeholder('float', [None, neurons_exit_layer])
```

```python
#Multi-Layer Perceptron!
def mlp(X, W, bias):
    #Mounting the tensorflow (graph of tensors)
    #According to: ((X . W) + bias)
    hidden_layer = tf.add( tf.matmul( X, W['hidden']), bias['hidden'] )
    
    #Using relu activation function
    hidden_layer_activation = tf.nn.relu(hidden_layer)
    #Exit layer receives the output of hidden layer as its input and calculate the summation (as function above)
    exit_layer = tf.add( tf.matmul( hidden_layer_activation, W['exit']), bias['exit'] )
    return exit_layer
```

```python
model = mlp(xph, W, b)
```

```python
#It is not necessary use softmax, cause the method below already do it!
error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits = model, labels = yph))
#Adam
optimizer = tf.train.AdamOptimizer(learning_rate = 0.0001).minimize(error)
```
